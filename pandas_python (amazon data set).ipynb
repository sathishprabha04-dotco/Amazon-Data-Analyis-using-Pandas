Perform initial data cleaning:

check missing values

```python
df.isnull().sum()
df.dropna(inplace=True)
```

remove duplicate rows

```python
df.drop_duplicates(inplace=True)
```

check column names

```python
print(df.columns)
```

clean the column (if any spaces)

```python
df.columns=df.columns.str.strip()
```

convert the column to numeric first:

```python
df['discounted_price']=(
df['discounted_price']
.replace('[₹,]','',regex=True)
.astype(float)
)
df['actual_price']=(
    df['actual_price']
    .replace('[₹,]','',regex=True)
    .astype(float)
)
df['discount_percentage']=(
    df['discount_percentage']
    .replace('%','',regex=True)
    .astype(float)
)
```

if rating is object/string:

```python
df['rating']=pd.to_numeric(df['rating'],errors='coerce')
```
